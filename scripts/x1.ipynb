```python
fund_dates = form13f_FF4["iPERIOD_END"].unique()
fund_dates
```

```python
fund_prices_FD = fund_prices_FF[fund_prices_FF["pSP_DATE"].isin(fund_dates)].reset_index()
fund_prices_FD
```

```python
form13f_FF4 = form13f_FF4.sort_values(['iPERIOD_END', 'iCIK'])
```

```python
fund_prices_FD['x1'] = 0
fund_prices_FD
```

```python
# 
for index, row in fund_prices_FD.iterrows():
    stock_id = row['pSP_CUSIP']
    date = row['pSP_DATE']
#     print(stock_id, date)
#     print(type(stock_id), type(date))
    stock_df = form13f_FF4[(form13f_FF4['iPERIOD_END'] == date) & (form13f_FF4['iCUSIP'] == stock_id)]
#     print(len(stock_df))
    fund_prices_FD.at[index, 'x1'] = len(stock_df)
```
