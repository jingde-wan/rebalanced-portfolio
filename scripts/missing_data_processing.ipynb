```python
data.x3.unique()
```

```python
len(data.x3.unique())
```

```python
data['x3'] = data['x3'].replace(['Small-Cap Blend', 'MidCap Blend', 'Large-Cap Growth',
                                 'MidCap Value', 'Small-Cap Growth', 'MidCap Growth',
                                 'Small-Cap Value', 'Large-Cap Blend', 'Large-Cap Value'],
                                [1,2,3,4,5,6,7,8,9])
data
```

```python
def performance_class(row):
    if row['y'] < -0.2:
        p = 0
    elif (row['y'] >= -0.2) and (row['y'] < -0.1):
        p = 1
    elif (row['y'] >= -0.1) and (row['y'] < -0.05):
        p = 2
    elif (row['y'] >= -0.05) and (row['y'] < 0):
        p = 3
    elif (row['y'] >= 0) and (row['y'] < 0.05):
        p = 4
    elif (row['y'] >= 0.05) and (row['y'] < 0.1):
        p = 5
    elif (row['y'] >= 0.1) and (row['y'] < 0.2):
        p = 6
    elif row['y'] >= 0.2:
        p = 7
    else:
        p = np.nan
    return p

data["class"] = data.apply(performance_class, axis=1)
data
```
