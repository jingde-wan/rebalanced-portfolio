## X15

The correlation between the past quarterly return of all good subfunds holding the stock, as defined by size/value cluster, and
the industry ETF

```python
x15['iPERIOD_END'].unique()
```

```python
x15 = good_funds[['iCUSIP','iPERIOD_END','subfund_vs','return_x','value_size_return']]
x15
```

```python
iCUSIP = []
iPERIOD = []
corr_vec = []
for p in x15['iPERIOD_END'].unique():
    for s in x15['iCUSIP'].unique():
        period_df = x15[x15['iPERIOD_END'] == p]
        stock_df = period_df[period_df['iCUSIP'] == s]
        corr = stock_df.iloc[:,3:].corr()['return_x']['value_size_return']
        corr_vec.append(corr)
        iCUSIP.append(s)
        iPERIOD.append(p)

x15 = pd.DataFrame({"iCUSIP": iCUSIP, "iPERIOD_END": iPERIOD, "x15": corr_vec})
```
