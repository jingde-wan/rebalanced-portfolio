```python
good_subfund_labels = pd.read_csv("good_subfund_labels.csv")
x2_3 = good_subfund_labels[['iCUSIP', 'iPERIOD_END', 'sector', 'value_size']]
x2_3.columns = ['iCUSIP', 'iPERIOD_END', 'x2', 'x3']
x2_3 = x2_3.sort_values(by='iCUSIP')
x2_3 = x2_3.drop_duplicates()
x2_3
```

```python
x67 = pd.read_csv('x67.csv')
x67 = x67.drop_duplicates()
x67 = x67[['iCUSIP','iPERIOD_END','x6','x7']]
x67 = x67.groupby(['iCUSIP','iPERIOD_END'],as_index=False)['x6','x7'].mean()
x67
```

```python
x8 = pd.read_csv('x8.csv')
x8 = x8.groupby(['iCUSIP', 'iPERIOD_END'], as_index=False).x8.mean()
x8
```

```python
x9_10_11 = pd.read_csv('x9-11.csv')
x9_10_11
```

```python
x12_13 = pd.read_csv('x12_13.csv')
x12_13 = x12_13.drop_duplicates()
x12_13 = x12_13[['iCUSIP','iPERIOD_END','x12','x13']]
x12_13 = x12_13.groupby(['iCUSIP','iPERIOD_END'],as_index=False)['x12','x13'].mean()
x12_13
```

```python
x19 = pd.read_csv('x19_new.csv')
x19 = x19[['pSP_CUSIP','pSP_DATE','X19_IDIO']]
x19.columns = ['iCUSIP','iPERIOD_END','x19']
x19
```

```python
x20_21 = pd.read_csv('funds_X20_X21.csv')
x20_21 = x20_21[['iCUSIP','iPERIOD_END','mar_Cap_change','quantity_change']]
x20_21.columns = ['iCUSIP','iPERIOD_END', 'x20', 'x21']
x20_21 = x20_21.groupby(['iCUSIP','iPERIOD_END'], as_index=False)[['x20','x21']].sum()
x20_21
```

```python
x16_18 = pd.read_csv('../../prices_X16_X17_X18.csv')
x16_18 = x16_18[['pSP_CUSIP','pSP_DATE','30D','60D','90D']]
x16_18.columns = ['iCUSIP','iPERIOD_END','x16','x17','x18']
x16_18
```

```python
y = pd.read_csv('return_target.csv')
y = y[['iCUSIP','iPERIOD_END','return']]
y.columns = ['iCUSIP','iPERIOD_END','y']
y
```

```python
dfs = [x2_3, x145, x67, x8, x9_10_11, x12_13, x14_15, x16_18, x19, x20_21, y]
```

```python
df_merged = reduce(lambda left,right: pd.merge(left,right,on=['iCUSIP','iPERIOD_END'], how='left'), dfs)
df_merged = df_merged[['iCUSIP','iPERIOD_END','x1','x2','x3','x4','x5','x6','x7','x8','x9','x10',
                       'x11','x12','x13','x14','x15','x16','x17','x18','x19','x20','x21','y']]
df_merged = df_merged.sort_values(by=['iPERIOD_END','iCUSIP'])
df_merged
```
